```python
import pandas as pd
import numpy as np
import re
```

```python
# KNN으로 채워넣은 수치값
df_rate = pd.read_csv('fillna_knn.csv')
df_rate = df_rate[['index', '단맛', '신맛', '청량감', '바디감', '도수']]
df_rate
```

```python
# 제품특징 기본 df
df_prod = pd.read_csv('traditional_liquor_df_final.csv')
df_prod.drop(columns = 'Unnamed: 0', axis = 1, inplace = True)
df_prod.drop(columns = ['단맛','신맛','청량감','바디감','도수%','탄산','주종','keyword'], inplace = True)
df_prod
```

```python
df_a = pd.merge(df_rate, df_prod, how = 'inner', on = 'index')
df_a
```

```python
# 키워드작업 결과물
df_keywords = pd.read_csv('keywords_final.csv')
df_keywords.drop(columns = ['제품명'], inplace = True)
df_keywords
```

```python
# 최종 df
df_final = pd.merge(df_a, df_keywords, how = 'inner', on = 'index')
df_final = df_final[['index','제품명','단맛','신맛','청량감','바디감','도수','용량','가격','keywords','제조사','원재료']]
df_final
```

```python
# 기존 키워드에 '저도수' 삭제
df_final['keywords'] = df_final['keywords'].apply(lambda x : re.sub(' 저도수',"", x))
```

```python
# 15% 이하인 제품 저도수 키워드 추가
df_final['저도'] = ""
for i in range(0,533) :
    if df_final.loc[i,'도수'] <= 15 :
        df_final.loc[i, '저도'] = " 저도수"
df_final['keywords'] = df_final['keywords'] + df_final['저도']
df_final.drop(columns = '저도', axis = 1, inplace = True)
df_final.rename(columns = {'keywords':'keyword'}, inplace = True)
```

```python
df_final
```

```python
# 저장
df_final.to_csv('df_final.csv', index = False)
```

```python
# 키워드기반 CBF 별도저장
df_cbf_database3 = df_final[['index','제품명','keyword']]
df_cbf_database3.to_csv('cbf_database3.csv', index = False)
```
